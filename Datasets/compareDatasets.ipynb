#### summarize each column

```python
def summarize_dataset(df):
    summary = []
    for column in df.columns:
        num_unique_values = df[column].nunique()
        
        col_data = {}
        col_data['Column Name'] = column
        col_data['Data Type'] = str(df[column].dtype)
        col_data['Total Number of Unique Values'] = num_unique_values
        
        # Skip writing Unique Values if there are more than 200 unique values
        if num_unique_values <= 200:
            col_data['Unique Values'] = df[column].unique().tolist()
        else:
            col_data['Unique Values'] = "Too many unique values to display"
        
        if df[column].dtype in ['int64', 'float64']:
            col_data['Statistics'] = {
                'Mean': df[column].mean(),
                'Min': df[column].min(),
                'Max': df[column].max(),
            }
        summary.append(col_data)

    # Convert the summary to a formatted string in Markdown format
    summary_text = ""
    for col in summary:
        summary_text += f"## {col['Column Name']}\n"
        summary_text += f"### Data Type: {col['Data Type']}\n"
        summary_text += f"### Total Number of Unique Values: {col['Total Number of Unique Values']}\n"
        summary_text += f"### Unique Values: {col['Unique Values']}\n"
        if 'Statistics' in col:
            stats = col['Statistics']
            summary_text += "### Statistics:\n"
            summary_text += f"#### Mean: {stats['Mean']}\n"
            summary_text += f"#### Min: {stats['Min']}\n"
            summary_text += f"#### Max: {stats['Max']}\n"
        summary_text += "\n"

    # Save the summary to a .md file with utf-8 encoding
    with open('BigVuln_dataset_summary.md', 'w', encoding='utf-8') as file:
        file.write(summary_text)

# Assuming you have your DataFrame named 'BigVuln'
summarize_dataset(BigVuln)
```

#### Summarize each column

```python
def summarize_dataset(df):
    summary = []
    for column in df.columns:
        try:
            num_unique_values = df[column].nunique()
        except TypeError:
            num_unique_values = 'N/A (contains unhashable types)'
        
        col_data = {}
        col_data['Column Name'] = column
        col_data['Data Type'] = str(df[column].dtype)
        col_data['Total Number of Unique Values'] = num_unique_values
        
        if isinstance(num_unique_values, int) and num_unique_values <= 200:
            col_data['Unique Values'] = df[column].unique().tolist()
        else:
            col_data['Unique Values'] = "Too many unique values to display" if isinstance(num_unique_values, int) else num_unique_values
        
        if df[column].dtype in ['int64', 'float64']:
            col_data['Statistics'] = {
                'Mean': df[column].mean(),
                'Min': df[column].min(),
                'Max': df[column].max(),
                'Std Dev': df[column].std()
            }
        summary.append(col_data)

    summary_text = ""
    for col in summary:
        summary_text += f"## {col['Column Name']}\n"
        summary_text += f"### Data Type: {col['Data Type']}\n"
        summary_text += f"### Total Number of Unique Values: {col['Total Number of Unique Values']}\n"
        summary_text += f"### Unique Values: {col['Unique Values']}\n"
        if 'Statistics' in col:
            stats = col['Statistics']
            summary_text += "### Statistics:\n"
            summary_text += f"#### Mean: {stats['Mean']}\n"
            summary_text += f"#### Min: {stats['Min']}\n"
            summary_text += f"#### Max: {stats['Max']}\n"
            summary_text += f"#### Std Dev: {stats['Std Dev']}\n"
        summary_text += "\n"

    with open('MegaVuln_dataset_summary.md', 'w', encoding='utf-8') as file:
        file.write(summary_text)

summarize_dataset(MegaVuln)
```

```python
hashes_megavuln = MegaVuln[['commit_hash']]
hashes_bigvuln = BigVuln[['commit_id']]
print("Total Number of hashes in MegaVuln :", len(hashes_megavuln)) 
print("Total Number of hashes in BigVuln :", len(hashes_bigvuln))

unique_megavuln = MegaVuln[['commit_hash']].drop_duplicates()
unique_bigvuln = BigVuln[['commit_id']].drop_duplicates()
print("Total Unique hashs in MegaVuln :",len(unique_megavuln))
print("Total Unique hashs in BigVuln :",len(unique_bigvuln))

intersected_rows = pd.merge(
    unique_megavuln,
    unique_bigvuln,
    left_on='commit_hash',
    right_on='commit_id'
)

num_intersected_rows = len(intersected_rows)

print(f"Number of intersected rows based on commit hash: {num_intersected_rows}")
```

### Vulnurable functions intersection

```python
vulnerable_megavuln = MegaVuln[MegaVuln['is_vul'] == True]
vulnerable_bigvuln = BigVuln[BigVuln['vul'] == 1]
print("Total Number of Vulnurable functions in MegaVuln :", len(vulnerable_megavuln)) 
print("Total Number of Vulnurable functions in BigVuln :", len(vulnerable_bigvuln))

vulnerable_megavuln = vulnerable_megavuln[['commit_hash']].drop_duplicates()
vulnerable_bigvuln = vulnerable_bigvuln[['commit_id']].drop_duplicates()
print("Total Unique hashs in MegaVuln :",len(vulnerable_megavuln))
print("Total Unique hashs in BigVuln :",len(vulnerable_bigvuln))

intersectionVuln = pd.merge(
    vulnerable_bigvuln,
    vulnerable_megavuln,
    left_on='commit_id',
    right_on='commit_hash'
)

num_intersected_rows = len(intersectionVuln)

print(f"Number of intersected rows based on commit hash: {num_intersected_rows}")
```

```python
vulnerable_megavuln = MegaVuln[MegaVuln['is_vul'] == True]
vulnerable_bigvuln = BigVuln[BigVuln['vul'] == 1]

print("Total Number of Vulnerable functions in MegaVuln:", len(vulnerable_megavuln))
print("Total Number of Vulnerable functions in BigVuln:", len(vulnerable_bigvuln))

megavuln_counts = vulnerable_megavuln['commit_hash'].value_counts()

bigvuln_counts = vulnerable_bigvuln['commit_id'].value_counts()

num_intersected_rows = 0

# Iterate over each commit hash in megavuln_counts and check if it exists in bigvuln_counts
for commit_hash in megavuln_counts.index:
    if commit_hash in bigvuln_counts:
        num_intersected_rows += min(megavuln_counts[commit_hash], bigvuln_counts[commit_hash])

print(f"Number of intersected rows vulnerable functions only (func_before) based on commit hash: {num_intersected_rows}")
```

```python
print("max Value of commit hash occurrences in MegaVuln",megavuln_counts.max())
print("Average  Value of commit hash occurrences in MegaVuln",megavuln_counts.mean())
print("min Value of commit hash occurrences in MegaVuln",megavuln_counts.min())
print("max Value of commit hash occurrences in BigVuln",bigvuln_counts.max())
print("Average  Value of commit hash occurrences in BigVuln",bigvuln_counts.mean())
print("min Value of commit hash occurrences in BigVuln",bigvuln_counts.min())
```

```python
# Get intersected commit hashes
intersected_commit_hashes = megavuln_counts.index.intersection(bigvuln_counts.index)

# Filter datasets for these commit hashes
intersected_megavuln = vulnerable_megavuln[vulnerable_megavuln['commit_hash'].isin(intersected_commit_hashes)]
intersected_bigvuln = vulnerable_bigvuln[vulnerable_bigvuln['commit_id'].isin(intersected_commit_hashes)]
print(len(intersected_megavuln))
print(len(intersected_bigvuln))
```

```python
vulnerable_megavuln = MegaVuln[MegaVuln['is_vul'] == True]
vulnerable_bigvuln = BigVuln[BigVuln['vul'] == 1]

print("Total Number of Vulnerable functions in MegaVuln:", len(vulnerable_megavuln))
print("Total Number of Vulnerable functions in BigVuln:", len(vulnerable_bigvuln))

# Count commit hash occurrences
megavuln_counts = vulnerable_megavuln['commit_hash'].value_counts()
bigvuln_counts = vulnerable_bigvuln['commit_id'].value_counts()

# Find intersecting commit hashes
intersected_commits = list(set(megavuln_counts.index).intersection(set(bigvuln_counts.index)))
print(f"Intersected Commits: {len(intersected_commits)}")

megavuln_indices = []
bigvuln_indices = []
num_intersected_rows = 0

for commit_hash in intersected_commits:
    # Get all indices (with duplicates) of the intersecting commit hash from both datasets
    mega_indices = vulnerable_megavuln[vulnerable_megavuln['commit_hash'] == commit_hash].index.tolist()
    big_indices = vulnerable_bigvuln[vulnerable_bigvuln['commit_id'] == commit_hash].index.tolist()
    
    # Store indices
    megavuln_indices.extend(mega_indices)
    bigvuln_indices.extend(big_indices)
    
    # Calculate the intersection count (min for each commit)
    num_intersected_rows += min(len(mega_indices), len(big_indices))

# Print results
print(f"Number of intersected rows vulnerable functions only (func_before) based on commit hash: {num_intersected_rows}")
print(f"Total intersected indices in MegaVuln: {len(megavuln_indices)}")
print(f"Total intersected indices in BigVuln: {len(bigvuln_indices)}")
```

```python
# Save the data to a file 
with open('intersected indices in MegaVuln.txt', 'w') as file:
    for index in megavuln_indices:
        file.write(f"{index}\n")

with open('intersected indices in BigVuln.txt', 'w') as file:
    for index in bigvuln_indices:
        file.write(f"{index}\n")
```
